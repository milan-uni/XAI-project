```python
from torchvision.models import vgg16
from torchvision import transforms
from PIL import Image
import torch
from explanations import RISE

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# Load VGG16 and set to eval mode
model = vgg16(pretrained=True).eval().cuda()

# Load and preprocess the bear image
img = Image.open('bear.jpg').convert('RGB')
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])
x = transform(img).unsqueeze(0)  # Shape: (1, 3, 224, 224)
```

```python
rise = RISE(model, input_size=(224, 224))
rise.generate_masks(N=3, s=7, p1=0.5) # N would be set to a high number like 8000 but google colab cuda is limited
```

```python
model.to(device)
x = x.to(device)
rise.masks = rise.masks.to(device)
```

```python
saliency = rise(x)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Suppose top predicted class is class_id
class_id = torch.argmax(model(x)).item()

# Normalize saliency map for visualization
sal_map = saliency[class_id].cpu().numpy()
sal_map = (sal_map - sal_map.min()) / (sal_map.max() - sal_map.min())

# Overlay on image
plt.imshow(img)
plt.imshow(sal_map, cmap='jet', alpha=0.5)
plt.title(f"Saliency for class {class_id}")
plt.axis('off')
plt.show()
```
